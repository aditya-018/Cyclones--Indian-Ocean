```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dt = pd.read_csv('cyc_bob.csv')
dt.head()
```

```python
dt = dt.drop('Unnamed: 0', axis=1)
dt
```

```python
dt.describe()
```

```python
dt.info()
```

```python
fig = plt.figure(figsize=(16,8))
plt.plot('Year', 'January - December (Whole Year): BOB', data=dt)
plt.xlabel('Year')
plt.ylabel('Freq')
plt.title('Cyclone Freq in BOB')
```

```python
fig = plt.figure(figsize=(16,8))
plt.plot('Year', 'January- February (winter): BOB', data=dt)
plt.xlabel('Year')
plt.ylabel('Freq')
```
